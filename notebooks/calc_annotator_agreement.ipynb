```python
from krippendorff import alpha
import pandas as pd
from ast import literal_eval
```

```python
ann1_df = pd.read_excel('annotation_results/Tejas/game_2024-03-09-14-44-25/annotation-2024-03-09-14-44-25.xlsx')
ann1_df.head()
```

```python
ann2_df = pd.read_csv('annotation_results/Arihant/game_2024-03-09-14-44-25/annotation-2024-03-09-14-44-25.csv')
ann2_df.head()
```

```python
ann2_df = ann2_df.dropna(subset=["intent_selection"])
ann2_df.head()
```

```python
len(ann1_df), len(ann2_df)
```

```python
ann1_df['intent'] = ann1_df['intent'].astype(str)
ann1_df['intent_mod'] = ann1_df['intent_mod'].astype(str)
ann2_df['intent'] = ann2_df['intent'].astype(str)
ann2_df['intent_mod'] = ann2_df['intent_mod'].astype(str)

ann1_df['intent'] = ann1_df['intent'].apply(lambda x: literal_eval(x))
ann1_df['intent_mod'] = ann1_df['intent_mod'].apply(lambda x: literal_eval(x))
ann2_df['intent'] = ann2_df['intent'].apply(lambda x: literal_eval(x))
ann2_df['intent_mod'] = ann2_df['intent_mod'].apply(lambda x: literal_eval(x))
```

```python
ann1_df['intent_selection'] = ann1_df['intent_selection'].astype(str)
ann1_df['intent_fol_think_1'] = ann1_df['intent_fol_think_1'].astype(str)
ann1_df['intent_fol_speech_1'] = ann1_df['intent_fol_speech_1'].astype(str)
ann1_df['intent_selection_mod'] = ann1_df['intent_selection_mod'].astype(str)
ann1_df['intent_fol_think_2'] = ann1_df['intent_fol_think_2'].astype(str)
ann1_df['intent_fol_speech_2'] = ann1_df['intent_fol_speech_2'].astype(str)

ann2_df['intent_selection'] = ann2_df['intent_selection'].astype(str)
ann2_df['intent_fol_think_1'] = ann2_df['intent_fol_think_1'].astype(str)
ann2_df['intent_fol_speech_1'] = ann2_df['intent_fol_speech_1'].astype(str)
ann2_df['intent_selection_mod'] = ann2_df['intent_selection_mod'].astype(str)
ann2_df['intent_fol_think_2'] = ann2_df['intent_fol_think_2'].astype(str)
ann2_df['intent_fol_speech_2'] = ann2_df['intent_fol_speech_2'].astype(str)
```

```python
ann1_df.head()
```

```python
ann_int_sel = {'ann1': {}, 'ann2': {}}
ann_int_fol1 = {'ann1': {}, 'ann2': {}}
ann_int_spk1 = {'ann1': {}, 'ann2': {}}
```

```python
def store_in_dic(row, ann):
    key = (row['player'], row['round'], row['is_secondary_round'])
    ann_int_sel[ann][key] = row['intent_selection'].split(',')
    ann_int_fol1[ann][key] = row['intent_fol_think_1'].split(',')
    ann_int_spk1[ann][key] = row['intent_fol_speech_1'].split(',')
```

```python
def store_in_dic_2(row, ann):
    key = (row['player'], row['round'], row['is_secondary_round'])
    ann_int_sel[ann][key] = row['intent_selection_mod'].split(',')
    ann_int_fol1[ann][key] = row['intent_fol_think_2'].split(',')
    ann_int_spk1[ann][key] = row['intent_fol_speech_2'].split(',')
```

```python
_ =  ann1_df.apply(lambda row: store_in_dic(row, 'ann1'), axis=1)
```

```python
_ =  ann2_df.apply(lambda row: store_in_dic(row, 'ann2'), axis=1)
```

```python
len(ann_int_fol1)
```

```python
len(ann_int_sel['ann1']), len(ann_int_sel['ann2'])
```

```python
int_sel_anns = []
for key, ann1_val in ann_int_sel['ann1'].items():
    ann2_val = ann_int_sel['ann2'][key]
    if ann1_val[0] == 'nan' or ann2_val[0] == 'nan':
        continue

    ann1_val_int = [int(x) for x in ann1_val]
    ann2_val_int = [int(x) for x in ann2_val]

    for x, y in zip(ann1_val_int, ann2_val_int):
        int_sel_anns.append([x, y])
```

```python
len(int_sel_anns)
```

```python
alpha_value = alpha(int_sel_anns)
print(f"Krippendorff's Alpha: {alpha_value:.3f}")
```
